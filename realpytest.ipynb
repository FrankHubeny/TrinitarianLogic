```python
mt = realpy.tf.Proof([A >> B, ~B], ~A, 'Modus Tollens')
mt.openblock(A)
mt.impelim(3,1)
mt.negelim(2,4)
mt.closeblock()
mt.negintro('11')
realpy.core.display(mt)
```

```python
t = realpy.tf.Proof([A, B], (A >> B) & (B >> A), 'Biconditional Introduction Test')
t.openblock(A)
t.reit(2)
t.closeblock()
t.impintro('11')
t.openblock(B)
t.reit(1)
t.closeblock()
t.impintro('12')
t.conjintro(8,5)
realpy.core.display(t)
```

```python
s = realpy.tf.Proof([(A >> B) & (B >> A)],A >> B,'Biconditional Elimination Test')
s.bicondelim(1)
realpy.core.display(s)
```

```python
r = realpy.tf.Proof([A & (B | C)], (A & B) | (A & C),'Trial')
r.conjelim(1)
r.openblock(B)
r.conjintro(2,4)
r.disjintro(A & C, 5)
r.closeblock()
r.openblock(C)
r.conjintro(2,7)
r.disjintro(A & B, 8)
r.closeblock()
r.disjelim(3, ['11','12'])
realpy.core.display(r)
```

```python
p = realpy.tf.Proof([A, A >> B], B, 'First Proof')
p.reit(1)
p.conjintro(1,2)
p.conjelim(4)
p.disjintro(C,1)
p.opensubproof(And(C, B))
p.conjelim(8)
p.conjintro(1,8)
p.conjelim(11)
p.closesubproof()
p.impintro('11')
#p.impelim(1,6)
#p.assumption(~A)
#p.negintro(3,17)

p.opensubproof(C)
realpy.core.display(p)
```
